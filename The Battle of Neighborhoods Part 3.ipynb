# Introduction/Business Problem

New York City is one of the world's most diverse cities to ever exist. Aside from a diverse demographic of cultures and people from all over the world, it has one of the largest transportation systems. Here in the big city, mass transit provides people with the opportunity to move around whether it be by foot, bus or subway. This is a key element that makes it ideal for opening a new restaurant within the five boroughs.
Mexican restaurants are already opened in New York City, although the city is huge, it becomes risky opening a restaurant of the same kind of foods within certain neighborhoods. This risk introduces factors such as, pricing, food quality and even popularity that can affect the performance of the new restaurant. Where would be the best location for opening a Mexican restaurant within New York City? What borough? What nieghborhoods?

# Data

I will obtain my data from this data source: https://cocl.us/new_york_dataset. This data includes the five boroughs and the neighborhoods within them. I will use FourSquare to filter and locate venues that are Mexican restaurants. I will use graphs to help assess what location will be ideal.

# Methodology

- I will collect data from https://cocl.us/new_york_dataset , then I will clean and move it into a dataframe
- I will use FourSquare to locate the venues and filter out the Mexican Restaurants
- I will take into account the total number of neighborhoods by borough
- I will also consider which borough has the least amount of Mexican restaurants
- Finally, I will consider which borough is ideal for opening a Mexican restaurant based off of number of neighborhoods and number of Mexican restaurants already opened within those neighborhoods

# Results

Based on the graph of Neighborhoods per Borough, Queens, Brooklyn and Staten Island have the most neighborhoods. Staten Island has 63 neighborhoods. However, when observing the graph Mexican Restaurants per Borough, Staten Island has the least amount of restaurants with there being 11 restaurants open in only 8 neighborhoods. 

# Discussion

Judging from the results, I would say the ideal location for opening a Mexican restaurant is in the borough of Staten Island. With this borough being the third borough with the most neighborhoods and having the least amount of Mexican restaurants, I would name this location to be the best. There is little to no room for competition and Staten Island has a lower real estate price rate compared to Manhattan and Brooklyn.
